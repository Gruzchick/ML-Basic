## Задание 1: Конвертер регистров

Написать функцию, которая будет переводить snake_case в PascalCase и наоборот. 

Функция должна сама определять - какой формат ей передали. Можно добавить ключевой аргумент, который будет принудительно возвращать один из форматов.

<br>

**Примеры:**
* `otus_course     -> OtusCourse`
* `PituhonIsTheBest -> pituhon_is_the_best`



```python
def convert_case(input_str):
    if not input_str:
        return ""

    result_chars = []
    capitalize_next = False

    if input_str[0].islower():
        result_chars.append(input_str[0].upper())
    else:
        result_chars.append(input_str[0].lower())

    for char in input_str[1:]:
        if char.isupper():
            result_chars.append('_')
            result_chars.append(char.lower())
        elif char == '_':
            capitalize_next = True
        elif capitalize_next:
            result_chars.append(char.upper())
            capitalize_next = False
        else:
            result_chars.append(char)

    return "".join(result_chars)

input_value = input()
result = convert_case(input_value)
print(result)
```

## Задание 2: Проверка валидности даты

Написать функцию проверяющую валидность введенной даты.

<br>

**Примеры:**
* `29.02.2000 -> True`
* `29.02.2001 -> False`
* `31.04.1962 -> False`



```python
def is_date_valid(date_str):
    try:
        day, month, year = map(int, date_str.split('.'))

        if not (1 <= month <= 12 and year > 0):
            return False
        elif month in (1, 3, 5, 7, 8, 10, 12):
            return 1 <= day <= 31
        elif month in (4, 6, 9, 11):
            return 1 <= day <= 30
        elif month == 2:
            is_leap = (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)
            if is_leap:
                return 1 <= day <= 29
            else:
                return 1 <= day <= 28
    except:
        return False

input_value = input()
result = is_date_valid(input_value)
print(result)
```

## Задание 3: Проверка на простое число

Функция проверки на простое число. Простые числа – это такие числа, которые делятся на себя и на единицу.

<br>

**Примеры:**
* `17 -> True`
* `20 -> False`
* `23 -> True`

```python
def is_prime(num):
    if num < 2:
        return False
    else:
        for i in range(2, int(num ** 0.5) + 1):
            if num % i == 0:
                return False
        return True

input_value = int(input())
result = is_prime(input_value)
print(result)
```

## Задание 4: Учет пользователей

Пользователь в бесконечном цикле вводит данные пользователей: имя, затем фамилию, возраст и ID. Ввод продолжается до тех пор, пока не будет введено пустое поле. 

Пользователи заносятся в словарь, где ключ это ID пользователя, а остальные данные записываются в виде кортежа. 

**Программа должна проверять:**
* имя и фамилия состоят только из символов и начинаются с большой буквы - если не с большой, то заменяет букву на большую;
* возраст должен быть числом от 18 до 60;
* ID - целое число, дополненное до 8 знаков незначащими нулями, ID должен быть уникальным.

**Дополнительно:** написать функцию, которая будет выводить полученный словарь в виде таблицы.

```python
def process_user_input(user_input, users_db):
    data = user_input.split()

    if len(data) != 4:
        print("Ошибка: Неверный формат ввода. Введите 4 значения, разделенных пробелом.")
        print("-" * 20)
        return

    name, last_name, age_str, user_id_str = data

    if not name.isalpha() or not last_name.isalpha():
        print("Ошибка: Имя и фамилия должны состоять только из букв.")
        print("-" * 20)
        return

    name = name.capitalize()
    last_name = last_name.capitalize()

    try:
        age = int(age_str)
        if not (18 <= age <= 60):
            print(f"Ошибка: Возраст '{age}' должен быть в диапазоне от 18 до 60.")
            print("-" * 20)
            return
    except ValueError:
        print(f"Ошибка: Возраст '{age_str}' должен быть целым числом.")
        print("-" * 20)
        return

    if not user_id_str.isdigit():
        print(f"Ошибка: ID '{user_id_str}' должен состоять только из цифр.")
        print("-" * 20)
        return

    formatted_id = user_id_str.zfill(8)

    if formatted_id in users_db:
        print(f"Ошибка: Пользователь с ID '{formatted_id}' уже существует.")
        print("-" * 20)
        return

    users_db[formatted_id] = (name, last_name, age)
    print(f"Пользователь {name} {last_name} успешно добавлен.")
    print("-" * 20)


users = {}

while True:
    user_input_str = input(
        "Введите данные (Имя Фамилия Возраст ID) или оставьте поле пустым для выхода: ")

    if not user_input_str:
        break

    process_user_input(user_input_str, users)

print("\nИтоговый список пользователей:")
for user_id, user_data in users.items():
    print(f"ID: {user_id}, Данные: {user_data}")
```
